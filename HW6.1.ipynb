```python
#Выполните агрегацию данных (например, вычислите среднее значение чисел в списке).

number = [-563, 24, 89, 3, 55.1, 211.11, -1, 99]
sum_num = sum(number)
avg = sum_num/len(number)
print("Среднее значение:")
print(avg)
print("Среднее значение списка до 3 десятичных знаком:")
print(round(avg,3))
```

```python
#Создайте график на основе данных с использованием библиотеки Matplotlib.

import matplotlib.pyplot as plt

labels = 'ОРВИ и грипп', 'Пневмания', 'Отравление', 'Травма'
sizes = [45, 30, 15, 10]

colors = ( "orange", "cyan", "brown",
          "grey")
fig1, ax1 = plt.subplots()
explode = (0.1, 0, 0, 0)
ax1.pie(sizes, colors = colors, explode=explode, labels=labels,autopct='%1.1f%%',
        shadow=True, startangle=90)
plt.title('Наиболее распространенные заболевания')
plt.show()
```
